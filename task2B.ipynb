```python
df_numeric = df.select_dtypes(include=[np.number])
numeric_cols = df_numeric.columns.values
print(numeric_cols)
```

```python
df_non_numeric = df.select_dtypes(exclude=[np.number])
non_numeric_cols = df_non_numeric.columns.values
print(non_numeric_cols)
```

***MISSING DATA HEAT MAP***

```python
cols = df.columns 
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(df[cols].isnull(), cmap=sns.color_palette(colours))
```

***MISSING DATA % LIST***

```python
df.isnull().sum()
```

```python
# if it's a larger dataset and the visualization takes too long can do this.
# % of missing.
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100,4)))
```

```python
df.shape
```

***REPLACING MISSING NUMERIC VALUES WITH MEDIAN***

```python
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.shape
```

```python
df[numeric_cols].isnull().sum()
```

***REPLACING NON NUMERIC DATA***

```python
for col in non_numeric_cols:
    df[col] = df[col].fillna(df[col].describe()['top'])
```

**WORLD POPULATION GROWTH RATE**

```python
world_total_population = df.loc[(df['Country Name'] == 'World') & (df['Series Name'] == 'Population, total' )]
world_total_population
```

```python
col_name = df.columns[4:]
years = []
for col in df.columns[4:]:
  years.append(col[:4])
print(col_name)
print(years)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(64, 16))
plt.plot(years, world_total_population[col_name].values.reshape(91,1))
plt.title('Annual world population growth rate',size = 40)
plt.xlabel('Year',size = 40)
plt.ylabel('Population',size = 40)
plt.show()
```

**WORLD POPULATION BY REGION**

```python
regions = df['Country Name'].unique()
print(regions)
print("No. of regions =",len(regions))
```

```python
plt.figure(figsize=(64, 80))
for reg in regions:
  plt.plot(years, df[df['Country Name']==reg].sum()[4:], label= reg)
plt.title('World population by region')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

**GROWTH OF THE POPULATION BY AGE GROUP**

```python
df3=pd.DataFrame(df.iloc[:,:4])
df3.rename(columns={'Country Name':'CName','Country Code':'CCode','Series Name':'SName','Series Code':'SCode'},inplace=True)
df4=pd.DataFrame(df.iloc[:,4:])
years=list(map(int,range(1960,2051)))
df4.columns=years
df_new = df3.join(df4, how='outer')
df_new
```

```python
df_SName=pd.DataFrame(df_new.iloc[:,2:])
df_SName.drop(['SCode'],axis=1,inplace=True)
df_SName_grouped=df_SName.groupby(['SName']).mean()
df_SName_grouped.reset_index(inplace=True)
df_SName_grouped.columns = list(map(str, df_SName_grouped.columns))
df_SName_grouped
```
